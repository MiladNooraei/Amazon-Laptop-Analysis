**Scraping each laptop's features**

```python
urls = unique_laptop_links_df["Link"].tolist()
all_product_details = []

def scrape_url(index_url_pair):
    index, url = index_url_pair
    retries = 3
    for attempt in range(retries):
        try:
            response = requests.get(url, timeout = 10)
            print(f"Fetching Laptop index: {index}, Attempt: {attempt + 1}, Status code: {response.status_code}")

            if response.status_code == 200:
                soup = BeautifulSoup(response.content, "html.parser")

                product_details = {}

                try:
                    price_tag = soup.find("span", {"class": "a-price"})
                    if price_tag and price_tag.find("span", {"class": "a-offscreen"}):
                        price_text = price_tag.find("span", {"class": "a-offscreen"}).get_text(strip = True).strip("$")
                        product_details["Price($)"] = float(price_text.replace(",", ""))
                    else:
                        product_details["Price($)"] = np.nan
                except Exception as e:
                    print(f"Error extracting price: {e}")
                    product_details["Price($)"] = np.nan

                table1 = soup.find("table", {"id": "productDetails_techSpec_section_1"})
                if table1:
                    for row in table1.find_all("tr"):
                        try:
                            feature_name = row.find("th").get_text(strip = True)
                            feature_value = row.find("td").get_text(strip = True)
                            product_details[feature_name] = feature_value
                        except AttributeError:
                            product_details[feature_name] = np.nan

                table2 = soup.find("table", {"id": "productDetails_techSpec_section_2"})
                if table2:
                    for row in table2.find_all("tr"):
                        try:
                            feature_name = row.find("th").get_text(strip = True)
                            feature_value = row.find("td").get_text(strip = True)
                            product_details[feature_name] = feature_value
                        except AttributeError:
                            product_details[feature_name] = np.nan

                product_details["Index"] = index
                return product_details

            else:
                print(f"Retrying... ({attempt + 1}/{retries})")
                time.sleep(1)

        except requests.exceptions.RequestException as e:
            print(f"RequestException for Laptop index: {index}, Attempt: {attempt + 1}, Error: {e}")
            if attempt == retries - 1:
                return None
            print(f"Retrying... ({attempt + 1}/{retries})")
            time.sleep(1)
    
    print(f"Fetching Laptop index: {index}, Failed to retrieve the page after {retries} attempts.")
    return None

with ThreadPoolExecutor(max_workers = 20) as executor:
    futures = list(executor.map(scrape_url, enumerate(urls, start = 1)))

all_product_details = [details for details in futures if details is not None]

laptops_df = pd.DataFrame(all_product_details)
```

**Reordering Columns**

```python
pd.set_option("display.max_columns", None)

new_columns_order = ["Index"] + [col for col in laptops_df.columns if col != "Index" and col != "Price($)"] + ["Price($)"]
laptops_df = laptops_df[new_columns_order]

laptops_df
```

**Converting to a CSV file**

```python
laptops_df.to_csv("laptops_dataset.csv", index = False)
```
